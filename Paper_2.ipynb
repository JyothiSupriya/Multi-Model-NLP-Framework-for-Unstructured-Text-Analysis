```python
from umap import UMAP
from hdbscan import HDBSCAN
from sentence_transformers import SentenceTransformer
from sklearn.feature_extraction.text import CountVectorizer
from bertopic import BERTopic
from bertopic.vectorizers import ClassTfidfTransformer
import nltk

# Download stopwords if not available
nltk.download('stopwords')
from nltk.corpus import stopwords

# Get the default English stopwords from NLTK and extend with additional custom stopwords
custom_stopwords = stopwords.words('english')
custom_stopwords.extend(["said", "say", "mr", "ms", "etc","nan", "nan see","see", "see attached", "attached"])

# Step 1 - Load the embedding model
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

# Step 2 - Reduce dimensionality of the embeddings
umap_model = UMAP(n_neighbors=15, n_components=5, min_dist=0.0, metric='cosine')

# Step 3 - Cluster reduced embeddings
hdbscan_model = HDBSCAN(min_cluster_size=5, metric='euclidean', cluster_selection_method='eom', prediction_data=True)

# Step 4 - Tokenize topics with manual stopword list
vectorizer_model = CountVectorizer(ngram_range=(1, 2),stop_words=custom_stopwords)

# Step 5 - Extract topic words using ClassTfidfTransformer
ctfidf_model = ClassTfidfTransformer()

# Step 6 - Create the BERTopic model
topic_model = BERTopic(
    embedding_model=embedding_model,
    umap_model=umap_model,
    hdbscan_model=hdbscan_model,
    vectorizer_model=vectorizer_model,        # Use custom stopwords
    ctfidf_model=ctfidf_model,calculate_probabilities=True
)

# Fit the model to your documents (replace `docs` with your actual text data)
'''docs = [
    "Air travel during COVID-19 is limited by new restrictions.",
    "Flights from India to Canada have been affected.",
    "Travel bans are preventing Australians from flying home.",
    "Aerosol transmission of COVID-19 in flights is a concern.",
    "The Airports Authority of India regulates flight operations.",
    "News about airport restrictions and flight cancellations due to COVID-19."
]'''
docs = [str(element) for element in docs]

topics, probs = topic_model.fit_transform(docs)

# Display related topics
related_topics = topic_model.find_topics("emergency", top_n=5)
print("Related topics to 'flight':", related_topics)

# Get the keywords for the related topics
for topic in related_topics[0]:
    print(f"\nTopic {topic}:")
    print(topic_model.get_topic(topic))
```

```python
import spacy

# Load spaCy model
nlp = spacy.load('en_core_web_sm')

def extract_root_cause(text):
    doc = nlp(text)
    root_causes = []

    # Define causal words and phrases
    causal_phrases = ['because', 'due to', 'caused by', 'as a result', 'resulting in', 'led to', 'leads to', 'caused', 'resulted in', 'triggered']

    # Iterate over sentences
    for sent in doc.sents:
        sent_text = sent.text.lower()
        if any(phrase in sent_text for phrase in causal_phrases):
            root_causes.append(sent.text.strip())
    return root_causes

# Example usage
text = """
ON JANUARY 29 2024, AROUND 4:40PM G3 UNDERGROUND (SUB OF ONYX CONSULTING WORKING FOR SPARKLIGHT), WAS PULLING BACK FIBER CONDUIT WITH A DIRECTIONAL DRILLING RIG AND DAMAGED 2" POLY MAIN WITH A MAKESHIFT ATTACHMENT THAT EXPANDS THE BORE HOLE THEY ORIGINALLY BORED. THE MAIN THAT WAS DAMAGED, WAS NOT POTHOLED PRIOR TO CONTRACTOR BORING. THE MAIN WAS PROPERLY AND CORRECTLY MARKED AND FLAGGED BY USIC. ESCAPING GAS FROM THE DAMAGED 2" POLY MAIN MIGRATED TO 502 AVE D RESULTING IN IGNITION OF GAS.
"""

root_causes = extract_root_cause(text)
print("Extracted Root Causes:")
for cause in root_causes:
    print(cause)

```

```python
import spacy
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer

# Load spaCy model
nlp = spacy.load('en_core_web_sm')

# Load BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForSequenceClassification.from_pretrained('bert-base-uncased')

def extract_root_cause(text):
    doc = nlp(text)
    root_causes = []

    # Iterate over sentences
    for sent in doc.sents:
        # Tokenize and classify each sentence
        inputs = tokenizer(sent.text, return_tensors='pt', truncation=True, padding=True)
        outputs = model(**inputs)
        logits = outputs.logits
        prediction = torch.argmax(logits, dim=-1).item()

        # Assuming label 1 indicates a causal relationship
        if prediction == 1:
            root_causes.append(sent.text.strip())
    return root_causes

def extract_topics(text_list):
    # Create BERTopic model
    vectorizer_model = CountVectorizer(ngram_range=(1, 2))
    topic_model = BERTopic(vectorizer_model=vectorizer_model)

    # Fit the model on the provided text list
    topics, probs = topic_model.fit_transform(text_list)
    return topic_model.get_topic_info()

# Example usage
text = """
ON JANUARY 29 2024, AROUND 4:40PM G3 UNDERGROUND (SUB OF ONYX CONSULTING WORKING FOR SPARKLIGHT), WAS PULLING BACK FIBER CONDUIT WITH A DIRECTIONAL DRILLING RIG AND DAMAGED 2" POLY MAIN WITH A MAKESHIFT ATTACHMENT THAT EXPANDS THE BORE HOLE THEY ORIGINALLY BORED. THE MAIN THAT WAS DAMAGED, WAS NOT POTHOLED PRIOR TO CONTRACTOR BORING. THE MAIN WAS PROPERLY AND CORRECTLY MARKED AND FLAGGED BY USIC. ESCAPING GAS FROM THE DAMAGED 2" POLY MAIN MIGRATED TO 502 AVE D RESULTING IN IGNITION OF GAS.
"""

root_causes = extract_root_cause(text)
print("Extracted Root Causes:")
for cause in root_causes:
    print(cause)

# Extract topics using BERTopic
text_list = [text]
topic_info = extract_topics(text_list)
print("\nExtracted Topics:")
print(topic_info)
```
